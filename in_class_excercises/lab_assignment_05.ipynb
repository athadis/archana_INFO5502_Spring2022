Question 3 (15 points). In this question, you are required to create visualization for time series data. 

We have collected all the articles from AMIA in the last 30 years. The dataset is avaliable on GitHub: https://raw.githubusercontent.com/unt-iialab/info5502-spring2022/main/datasets/lab_assignment_05_amia_cleaned.csv. 

Please use the tool Yake: https://github.com/LIAAD/yake to extract the top N (you can set different N value to get the best visualization results) keywords from the article titles (all the articles), then visualize how the keywords changed over years in one figure. (Split the data by every five years)

```python
!pip install git+https://github.com/LIAAD/yake
from numpy.ma.core import maximum
import pandas
import yake
import matplotlib.pylab as plt

extracted_data=pandas.read_csv('/content/lab_assignment_05_amia_cleaned.csv')

#sort data in ascending order by year
sorted_df = extracted_data.sort_values(["year"], ascending=True)

#selected all rows that have a valid year values
sorted_df = sorted_df[sorted_df['year'].notna()]

#Pass concatenated title for every 5 years to yake
#store the output in a dictionary 
custom_kw_extractor = yake.KeywordExtractor(top=5, features=None)

result= pandas.DataFrame(data=None,columns=["Keyword","Score","Year"])
max_year=int(sorted_df['year'].max())
min_year=int(sorted_df['year'].min())

while min_year<=max_year:
  filtered_df=sorted_df[sorted_df['year'].isin([min_year,min_year+1,min_year+2,min_year+3,min_year+4])]
  text = ' '.join(filtered_df["title"])
  keywords = custom_kw_extractor.extract_keywords(text)
  result.update({min_year:keywords})
  df= pandas.DataFrame(data=keywords,columns=["Keyword","Score"])
  df["Year"]=min_year
  result= pandas.concat([result,df], ignore_index=True)
  min_year=min_year+5

fig, ax = plt.subplots(figsize=(15,10))
for key, grp in result.groupby(['Keyword']):
    ax.plot(grp['Year'], grp['Score'], label=key,marker='+')

ax.legend()
plt.title('Top 5 search terms data by Year')
plt.xlabel('Year')
plt.ylabel('Score')
plt.show()
```
